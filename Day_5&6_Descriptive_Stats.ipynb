```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("k_circle_sales.csv")
```

```python
df
```

```python
# A) Number of rows and columns
print("Shape of dataset:")
print(df.shape)

# B) Datatypes
print("\nDatatypes:")
print(df.dtypes)

# C) Columns
print("\nColumns:")
print(df.columns)

# D) Description
print("\nStatistical Description:")
print(df.describe(include='all'))
```

```python
# 1. Count values in item_fat_content column
# =========================================================

print("\nCount of values in item_fat_content:")
print(df['Item_Fat_Content'].value_counts())
```

```python
# 2. Find unique values in item_identifier
# =========================================================

print("\nUnique values in item_identifier:")
print(df['Item_Identifier'].unique())
```

```python
# 3. Replace LF with Low Fat and reg with Regular
# =========================================================

df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({
    'LF': 'Low Fat',
    'reg': 'Regular'
})

print("\nUpdated item_fat_content values:")
print(df['Item_Fat_Content'].value_counts())
```

```python
# 4. Find mean weight of all soft drinks
# =========================================================

soft_drinks = df[df['Item_Type'] == 'Soft Drinks']

print("\nMean weight of Soft Drinks:")
print(soft_drinks['Item_Weight'].mean())
```

```python
# 5. Find mean weight of items whose profit is more than 10.5
# =========================================================

profit_items = df[df['Profit'] > 10.5]

print("\nMean weight of items with profit > 10.5:")
print(profit_items['Item_Weight'].mean())
```

```python
# 6. Find IQR of item_MRP
# =========================================================

Q1 = df['Item_MRP'].quantile(0.25)
Q3 = df['Item_MRP'].quantile(0.75)

IQR = Q3 - Q1

print("\nIQR of item_MRP:")
print(IQR)
```

```python
# 7. Find the 30th percentile of item_MRP
# =========================================================

percentile_30 = df['Item_MRP'].quantile(0.30)

print("\n30th Percentile of item_MRP:")
print(percentile_30)
```

```python
# 8. Variance, Standard Deviation, Coefficient of Variation
# =========================================================

variance = df['Item_MRP'].var()
std_dev = df['Item_MRP'].std()
mean_mrp = df['Item_MRP'].mean()

cv = (std_dev / mean_mrp) * 100

print("\nVariance of item_MRP:")
print(variance)

print("\nStandard Deviation of item_MRP:")
print(std_dev)

print("\nCoefficient of Variation of item_MRP:")
print(cv)
```

```python
# 9. Skewness and Kurtosis of item_identifier
# =========================================================
# item_identifier is usually categorical/string,
# so skewness and kurtosis are not applicable directly.
# If needed, convert to categorical codes.

identifier_codes = df['Item_Identifier'].astype('category').cat.codes

print("\nSkewness of item_identifier:")
print(identifier_codes.skew())

print("\nKurtosis of item_identifier:")
print(identifier_codes.kurt())
```

```python
# 10. Histogram for item_type
# =========================================================

plt.figure(figsize=(12,6))

df['Item_Type'].value_counts().plot(kind='bar')

plt.title("Histogram / Bar Graph of Item Type")
plt.xlabel("Item Type")
plt.ylabel("Count")

plt.xticks(rotation=90)

plt.show()
```
