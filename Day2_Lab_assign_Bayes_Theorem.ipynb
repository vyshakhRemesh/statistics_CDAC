# **Ques 1:**
A factory has two machines. Machine A produces 70% of the items and has a 4% defect rate. Machine B produces 30% of the items and has a 5% defect rate. If a randomly selected item is found to be defective,what is the probability that it was produced by Machine A?(0.6512)

```


p_a=0.7
p_b=0.3
p_d_a=0.04
p_d_b=0.05

p_a_d=(p_a*p_d_a)/(p_a*p_d_a+p_b*p_d_b)
print(p_a_d)
```

# **Ques 2:**
A disease affects 2% of the population. A medical test detects disease correctly by 95% and false positive rate is 10%.Find the probability that a person actually has disease if test result is positive.(0.162)

```


p_d=0.02
p_nd=1-p_d
p_pos_d=0.95
p_pos_nd=0.1
p_d_pos=(p_d*p_pos_d)/(p_d*p_pos_d+p_nd*p_pos_nd)
print(p_d_pos)

```


# **Ques 3:**
A company has three suppliers. Supplier A supplies 50%of the parts, with 2% defective. Supplier B supplies 30%of the parts, with 3% defective. Supplier C supplies 20%of the parts, with 5% defective. From the total production, a part is selected at random and is found to be defective.What is the probability that the defective part came from Supplier C?(0.345)

```
p_a=0.5
p_d_a=0.02

p_b=0.3
p_d_b=0.03

p_c=0.2
p_d_c=0.05

p_c_d=(p_c*p_d_c)/(p_c*p_d_c+p_a*p_d_a+p_b*p_d_b)
print(p_c_d)
```
