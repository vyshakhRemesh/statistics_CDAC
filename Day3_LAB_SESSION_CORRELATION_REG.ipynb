**Q1: Find the covariance and correlation betweenItemsPurchased and Total Spendand interpret the result**

```python
x=df["Items Purchased"]
y=df["Total Spend"]
correlation = x.corr(y)
print(correlation)
```

```python
covariance=np.cov(x,y)
print(covariance)
```

**Q6: Create a dataframe for values X = [10, 20, 30, 40, 1000] and Y = [15, 25, 35, 45, 1100]**

```python
df_new=pd.DataFrame(
    {
        "x":[10, 20, 30, 40, 1000],
        "y":[15, 25, 35, 45, 1100]
    }
)
df_new
```

a) Calculate covariance and correlation between X and Y

```python
correlation=df_new["x"].corr(df_new["y"])
print(correlation)
```

```python
covariance=df_new["x"].cov(df_new["y"])
print(covariance)
```

b) Check for outliers in X and Y and remove them

```python
Q1=df_new["x"].quantile(0.25)
Q3=df_new["y"].quantile(0.75)

IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR

print(lower,upper)
```

```python
clean_data=df_new[(df_new["x"]>=lower)&
              (df_new["y"]<=upper)]
print(clean_data.shape)
print(df_new.shape)
```

c) Recalculate the covariance and correlation between X and Y after removing outliers.

```python
new_corr=clean_data["x"].corr(clean_data["y"])
new_cov=clean_data["x"].cov(clean_data["y"])

print(new_corr,new_cov)
```
